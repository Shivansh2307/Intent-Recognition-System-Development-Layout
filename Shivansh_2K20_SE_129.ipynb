```python
import numpy as np  # Numpy is used for efficient numerical computations and data manipulation essential for preprocessing the dataset.
import tensorflow as tf  # TensorFlow is used to build, train, and deploy the deep learning model.
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.preprocessing import LabelEncoder
import json

# This is a sample dataset
data = {
    "intents": [
        {"text": "Hello, how can I help you?", "intent": "greeting"},
        {"text": "Hello", "intent": "greeting"},
        {"text": "What is the weather today?", "intent": "weather"},
        {"text": "Book a flight for me", "intent": "booking"},
        {"text": "Thank you!", "intent": "thanks"},
        {"text": "Goodbye", "intent": "goodbye"},
        # We can add more data here or can replace it with an actual big dataset as well.
    ]
}

# Extracting texts and labels
texts = [item['text'] for item in data['intents']]
labels = [item['intent'] for item in data['intents']]

# Tokenizing texts (breaking down into tokens)
tokenizer = Tokenizer(num_words=1000, oov_token="<OOV>")
tokenizer.fit_on_texts(texts)
sequences = tokenizer.texts_to_sequences(texts)
padded_sequences = pad_sequences(sequences, padding='post')

# Encoding labels
label_encoder = LabelEncoder()
label_encoder.fit(labels)
encoded_labels = label_encoder.transform(labels)
# This is used to convert categorical labels into numerical values, making them compatible with machine learning models.

# Building our basic model
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(input_dim=1000, output_dim=16, input_length=padded_sequences.shape[1]),
    tf.keras.layers.GlobalAveragePooling1D(),
    tf.keras.layers.Dense(16, activation='relu'),  # relu is used in hidden layers because of its efficiency in learning complex patterns.
    tf.keras.layers.Dense(len(set(labels)), activation='softmax')  # softmax is used in the outer layer for its ability to handle multiple cases and multi-class classification.
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Training our model
model.fit(padded_sequences, np.array(encoded_labels), epochs=500)

# Saved the model and tokenizer
model.save('intent_model.h5')
with open('tokenizer.json', 'w') as f:
    f.write(json.dumps(tokenizer.to_json()))

# This is our function to predict intent
def predict_intent(text):
    seq = tokenizer.texts_to_sequences([text])
    padded_seq = pad_sequences(seq, maxlen=padded_sequences.shape[1])
    prediction = model.predict(padded_seq)
    return label_encoder.inverse_transform([np.argmax(prediction)])[0]

# Testing the prediction
print(predict_intent("Hello"))  #This Should output 'greeting'

```
